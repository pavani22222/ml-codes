```python
import math

def euclidean_distance(x, y):
    """
    Calculates the Euclidean distance between two points.
    """
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(x, y)))

def manhattan_distance(x, y):
    """
    Calculates the Manhattan distance between two points.
    """
    return sum(abs(a - b) for a, b in zip(x, y))

def minkowski_distance(x, y, p):
    """
    Calculates the Minkowski distance between two points.
    """
    return sum(abs(a - b) ** p for a, b in zip(x, y)) ** (1/p)
point1 = input("Enter coordinates of the first point (comma separated): ").split(',')
point2 = input("Enter coordinates of the second point (comma separated): ").split(',')
point1 = [float(coord) for coord in point1]
point2 = [float(coord) for coord in point2]
euclidean_dist = euclidean_distance(point1, point2)
manhattan_dist = manhattan_distance(point1, point2)
k = 3 
minkowski_dist = minkowski_distance(point1, point2, k)
print("Euclidean distance:", euclidean_dist)
print("Manhattan distance:", manhattan_dist)
print("Minkowski distance with k =", k, ":", minkowski_dist)
```

```python
from sklearn.datasets import load_iris
from sklearn.metrics.pairwise import euclidean_distances
iris = load_iris()
data = iris.data
distance_matrix = euclidean_distances(data)
print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
data = iris.data
target = iris.target
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target
class1 = 0  
class2 = 1
class1_data = df[df['target'] == class1].iloc[:, :2]  
class2_data = df[df['target'] == class2].iloc[:, :2]  
plt.scatter(class1_data.iloc[:, 0], class1_data.iloc[:, 1], label=f'Class {class1}')
plt.scatter(class2_data.iloc[:, 0], class2_data.iloc[:, 1], label=f'Class {class2}')
plt.xlabel(iris.feature_names[0])  
plt.ylabel(iris.feature_names[1])  
plt.legend()
plt.title("Scatter Plot of Two Classes")
plt.show()
```
