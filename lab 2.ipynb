```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
title = "Sample Plot"
x_label = "X-axis"
y_label = "Y-axis"
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)
ax.set_title(title)
plt.show()
```

```python
import matplotlib.pyplot as plt
with open('data.txt', 'r') as file:
    data = file.readlines()
X = []
Y = []
for line in data:
    x, y = map(float, line.split())
    X.append(x)
    Y.append(y)
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Line Plot from Data File')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]
plt.plot(x, y1, label='Line 1', color='blue', linewidth=2)
plt.plot(x, y2, label='Line 2', color='red', linewidth=3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two Lines with Legends')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x_positions = [1, 2, 3, 4, 5]
y_positions = [5, 7, 2, 8, 4]
plt.figure(figsize=(8, 6)) 
plt.scatter(x_positions, y_positions, color='blue', marker='o')  
plt.title('Quantities with X and Y Positions')  
plt.xlabel('X Position')  
plt.ylabel('Y Position')  
plt.grid(True) 
plt.show() 
```

```python
import matplotlib.pyplot as plt
x_positions1 = [1, 2, 3, 4, 5]
y_positions1 = [5, 7, 2, 8, 4]
x_positions2 = [2, 3, 4, 5, 6]
y_positions2 = [8, 6, 3, 5, 2]
fig, axs = plt.subplots(2, 1, figsize=(8, 10))
axs[0].scatter(x_positions1, y_positions1, color='blue', marker='o')
axs[0].set_title('First Set of Quantities')
axs[0].set_xlabel('X Position')
axs[0].set_ylabel('Y Position')
axs[0].grid(True)
axs[1].scatter(x_positions2, y_positions2, color='red', marker='x')
axs[1].set_title('Second Set of Quantities')
axs[1].set_xlabel('X Position')
axs[1].set_ylabel('Y Position')
axs[1].grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for x and y positions
num_points = 100
x_positions = np.random.rand(num_points)  # Generate 100 random x positions between 0 and 1
y_positions = np.random.rand(num_points)  # Generate 100 random y positions between 0 and 1

# Create scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(x_positions, y_positions, color='green', marker='o', alpha=0.7)  # Scatter plot
plt.title('Random Data Scatter Plot')  # Title of the plot
plt.xlabel('X Position')  # X-axis label
plt.ylabel('Y Position')  # Y-axis label
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```

```python
import matplotlib.pyplot as plt

# Sample data: product names and their respective sales
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [350, 500, 400, 600]  # Sales figures for each product

# Create a bar chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(products, sales, color='skyblue')  # Create the bar chart
plt.xlabel('Products')  # Label for the x-axis
plt.ylabel('Sales (units)')  # Label for the y-axis
plt.title('Sales of Different Products')  # Title of the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines to the y-axis
plt.show()  # Show the bar chart
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['Housing', 'Transportation', 'Food', 'Utilities', 'Entertainment', 'Healthcare']
expenses = [1500, 600, 800, 300, 400, 200]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink', 'lightgrey']
plt.figure(figsize=(8, 8))
plt.pie(expenses, labels=categories, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Expenses in Budget')
plt.axis('equal')
plt.show()
```
